***3.Explain the following terms***

1. The difference in np.array, np.asarray and np.asanyarray.
2. The difference between Deep copy and shallow copy

Ans. 1.np.array()

- Creates a new NumPy array object, copying data if necessary.
- Ensures a contiguous memory block.
- Allows specifying dtype, shape, and order.



2. np.asarray:
    - Creates a view of the input data if possible (no copying).
    - If input is already an array, returns the original array (no copying).
    - Otherwise, creates a new array object, copying data.
    - Preserves the original dtype and shape.
    - Returns a view or a new array object.
3. np.asanyarray:
    - Similar to np.asarray, but:
    - Always returns a new array object (never a view).
    - Copies data if necessary, even if input is already an array.
    - Ensures a contiguous memory block.



2. The difference between Deep copy and shallow copy


Shallow Copy

A shallow copy constructs a new compound object and then (to the extent possible) inserts references into it to the original elements.

Deep Copy

A deep copy constructs a new compound object and then, recursively, inserts copies into it of the objects found in the original.

***14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:***

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
# Create a time-series dataset
dates = pd.date_range('2022-01-01', periods=12, freq='M')
temperature = np.random.uniform(20, 40, size=12)
humidity = np.random.uniform(50, 90, size=12)

```

```python
df
```

```python
# Plot the 'Temperature' and 'Humidity' on the same plot
fig, ax1 = plt.subplots()

ax1.plot(df['Date'], df['Temperature'], color='red')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.title('Temperature and Humidity Over Time')

# Display a legend
line1, = ax1.plot(df['Date'], df['Temperature'], color='red')
line2, = ax2.plot(df['Date'], df['Humidity'], color='blue')
plt.legend([line1, line2], ['Temperature', 'Humidity'], loc='upper right')

plt.show()
```

***15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:***

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
mean, std_dev = 0, 1
data = np.random.normal(mean, std_dev, 1000)
```

```python
# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, label='Histogram')

# Label the axes and set the title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Overlay a line plot representing the normal distribution's probability density function (PDF)
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
pdf = norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, color='red', label='Normal Distribution PDF')

# Display the plot
plt.show()

```

***16.  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.***

```python
import seaborn as sns

# Create two random arrays
np.random.seed(0)
x = np.random.uniform(-10, 10, size=100)
y = np.random.uniform(-10, 10, size=100)
```

```python
# Define quadrants
quadrant = []
for i, j in zip(x, y):
    if i >= 0 and j >= 0:
        quadrant.append('I')
    elif i < 0 and j >= 0:
        quadrant.append('II')
    elif i < 0 and j < 0:
        quadrant.append('III')
    else:
        quadrant.append('IV')
```

```python
# Create a Seaborn scatter plot
plt.figure(figsize=(8, 8))
sns.scatterplot(x=x, y=y, hue=quadrant, palette=['blue', 'green', 'red', 'yellow'], hue_order=['I', 'II', 'III', 'IV'])

# Add a legend
plt.legend(title='Quadrants', bbox_to_anchor=(1.05, 1), loc='upper left')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Set aspect ratio to 'equal' for symmetric plot
plt.gca().set_aspect('equal')

# Display the plot
plt.show()


```

***19. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'.

```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.random.randn(100) + x

# Create figure
fig = go.Figure(data=[go.Scatter(x=x, y=y)])

# Customize plot
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
)

# Display plot
fig.show()
```
